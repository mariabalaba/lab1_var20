# 2.4.7
Напишите программу на Python, чтобы нарисовать точечный график, чтобы обнаружить повышение уровня моря за последние 100 лет.  
Фрагмент кода дает вывод, показанный на следующем снимке экрана:
image.png

```
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('classic')
data = pd.read_csv('data.csv')
Time = data['Time']
data['GMSL (mm)'] = data['GMSL (mm)'] / 25.4 
sea_levels = data['GMSL (mm)']
plt.scatter(Time, sea_levels, edgecolors='g', color = 'b', alpha = 0.6)
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sealevel')
plt.show()
```
